```python
#Importing packages
import rdflib as rdf
from rdflib import Graph,Literal,RDF
from rdflib.namespace import RDFS,XSD


class App:

    #Function to create the tbox from input empty graph
    def tbox(self, graph):
        
        #Defining namespace
        URI = "http://sdm_lab/"
        sdm_lab = rdf.Namespace(URI)
        #Bind prefix to the namespace we created using our URI
        graph.bind("sdm_lab",sdm_lab)
        
        ################## Classes ######################
        graph.add((sdm_lab.Paper, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Paper, RDFS.label, Literal("Paper")))
        
        graph.add((sdm_lab.Keyword, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Keyword, RDFS.label, Literal("Keyword")))
        
        graph.add((sdm_lab.Proceeding, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Proceeding, RDFS.label, Literal("Proceeding")))
        
        graph.add((sdm_lab.Venue, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Venue, RDFS.label, Literal("Venue")))
        
        graph.add((sdm_lab.Volume, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Volume, RDFS.label, Literal("Volume")))
        
        graph.add((sdm_lab.Responsible, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Responsible, RDFS.label, Literal("Responsible")))
        
        graph.add((sdm_lab.Author, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Author, RDFS.label, Literal("Author")))
        
        graph.add((sdm_lab.Review, RDF.type, RDFS.Class))
        graph.add((sdm_lab.Review, RDFS.label, Literal("Review")))
        
        ################## Properties ######################
        graph.add((sdm_lab.has, RDFS.domain, sdm_lab.Conference))
        graph.add((sdm_lab.has, RDFS.range, sdm_lab.Proceeding))
        graph.add((sdm_lab.has, RDFS.label, Literal("has")))

        graph.add((sdm_lab.cHandledBy, RDFS.domain, sdm_lab.Conference))
        graph.add((sdm_lab.cHandledBy, RDFS.range, sdm_lab.Chair))
        graph.add((sdm_lab.cHandledBy, RDFS.label, Literal("cHandledBy")))

        graph.add((sdm_lab.jHandledBy, RDFS.domain, sdm_lab.Journal))
        graph.add((sdm_lab.jHandledBy, RDFS.range, sdm_lab.Editor))
        graph.add((sdm_lab.jHandledBy, RDFS.label, Literal("jHandledBy")))

        graph.add((sdm_lab.cRelatedTo, RDFS.domain, sdm_lab.Conference))
        graph.add((sdm_lab.cRelatedTo, RDFS.range, sdm_lab.Keyword))
        graph.add((sdm_lab.cRelatedTo, RDFS.label, Literal("cRelatedTo")))

        graph.add((sdm_lab.jRelatedTo, RDFS.domain, sdm_lab.Journal))
        graph.add((sdm_lab.jRelatedTo, RDFS.range, sdm_lab.Keyword))
        graph.add((sdm_lab.jRelatedTo, RDFS.label, Literal("jRelatedTo")))

        graph.add((sdm_lab.pRelatedTo, RDFS.domain, sdm_lab.Paper))
        graph.add((sdm_lab.pRelatedTo, RDFS.range, sdm_lab.Keyword))
        graph.add((sdm_lab.pRelatedTo, RDFS.label, Literal("pRelatedTo")))

        graph.add((sdm_lab.isSubmittedTo, RDFS.domain, sdm_lab.Paper))
        graph.add((sdm_lab.isSubmittedTo, RDFS.range, sdm_lab.Venue))
        graph.add((sdm_lab.isSubmittedTo, RDFS.label, Literal("isSubmittedTo")))

        graph.add((sdm_lab.word, RDFS.domain, sdm_lab.Keyword))
        graph.add((sdm_lab.word, RDFS.range, XSD.string))
        graph.add((sdm_lab.word, RDFS.label, Literal("word")))

        graph.add((sdm_lab.contains, RDFS.domain, sdm_lab.Journal))
        graph.add((sdm_lab.contains, RDFS.range, sdm_lab.Volume))
        graph.add((sdm_lab.contains, RDFS.label, Literal("contains")))

        graph.add((sdm_lab.assigns, RDFS.domain, sdm_lab.Responsible))
        graph.add((sdm_lab.assigns, RDFS.range, sdm_lab.Reviewer))
        graph.add((sdm_lab.assigns, RDFS.label, Literal("assigns")))

        graph.add((sdm_lab.performs, RDFS.domain, sdm_lab.Reviewer))
        graph.add((sdm_lab.performs, RDFS.range, sdm_lab.Review))
        graph.add((sdm_lab.performs, RDFS.label, Literal("performs")))

        graph.add((sdm_lab.reviewText, RDFS.domain, sdm_lab.Review))
        graph.add((sdm_lab.reviewText, RDFS.range, XSD.string))
        graph.add((sdm_lab.reviewText, RDFS.label, Literal("reviewText")))
        
        graph.add((sdm_lab.reviewDecision, RDFS.domain, sdm_lab.Review))
        graph.add((sdm_lab.reviewDecision, RDFS.range, XSD.string))
        graph.add((sdm_lab.reviewDecision, RDFS.label, Literal("reviewDecison")))

        graph.add((sdm_lab.isUnderReview, RDFS.domain, sdm_lab.Paper))
        graph.add((sdm_lab.isUnderReview, RDFS.range, sdm_lab.Review))
        graph.add((sdm_lab.isUnderReview, RDFS.label, Literal("isUnderReview")))

        graph.add((sdm_lab.name, RDFS.domain, sdm_lab.Author))
        graph.add((sdm_lab.name, RDFS.range, XSD.string))
        graph.add((sdm_lab.name, RDFS.label, Literal("name")))

        graph.add((sdm_lab.writes, RDFS.domain, sdm_lab.Author))
        graph.add((sdm_lab.writes, RDFS.range, sdm_lab.Paper))
        graph.add((sdm_lab.writes, RDFS.label, Literal("writes")))

        graph.add((sdm_lab.title, RDFS.domain, sdm_lab.Paper))
        graph.add((sdm_lab.title, RDFS.range, XSD.string))
        graph.add((sdm_lab.title, RDFS.label, Literal("title")))

        graph.add((sdm_lab.year, RDFS.domain, sdm_lab.Paper))
        graph.add((sdm_lab.year, RDFS.range, XSD.integer))
        graph.add((sdm_lab.year, RDFS.label, Literal("year")))

        ################## Subclasses ######################
        #Conference
        graph.add((sdm_lab.Workshop, RDFS.subClassOf, sdm_lab.Conference))
        graph.add((sdm_lab.Workshop, RDFS.label, Literal("Workshop")))

        graph.add((sdm_lab.Symposium, RDFS.subClassOf, sdm_lab.Conference))
        graph.add((sdm_lab.Symposium, RDFS.label, Literal("Symposium")))

        graph.add((sdm_lab.ExpertGroup, RDFS.subClassOf, sdm_lab.Conference))
        graph.add((sdm_lab.ExpertGroup, RDFS.label, Literal("ExpertGroup")))

        graph.add((sdm_lab.RegularConference, RDFS.subClassOf, sdm_lab.Conference))
        graph.add((sdm_lab.RegularConference, RDFS.label, Literal("Workshop")))

        #Paper
        graph.add((sdm_lab.FullPaper, RDFS.subClassOf, sdm_lab.Paper))
        graph.add((sdm_lab.FullPaper, RDFS.label, Literal("FullPaper")))

        graph.add((sdm_lab.ShortPaper, RDFS.subClassOf, sdm_lab.Paper))
        graph.add((sdm_lab.ShortPaper, RDFS.label, Literal("ShortPaper")))

        graph.add((sdm_lab.DemoPaper, RDFS.subClassOf, sdm_lab.Paper))
        graph.add((sdm_lab.DemoPaper, RDFS.label, Literal("DemoPaper")))

        graph.add((sdm_lab.Poster, RDFS.subClassOf, sdm_lab.Paper))
        graph.add((sdm_lab.Poster, RDFS.label, Literal("Poster")))

        #Venue
        graph.add((sdm_lab.Conference, RDFS.subClassOf, sdm_lab.Venue))
        graph.add((sdm_lab.Conference, RDFS.label, Literal("Conference")))

        graph.add((sdm_lab.Journal, RDFS.subClassOf, sdm_lab.Venue))
        graph.add((sdm_lab.Journal, RDFS.label, Literal("Journal")))

        #Responsible
        graph.add((sdm_lab.Chair, RDFS.subClassOf, sdm_lab.Responsible))
        graph.add((sdm_lab.Chair, RDFS.label, Literal("Chair")))

        graph.add((sdm_lab.Editor, RDFS.subClassOf, sdm_lab.Responsible))
        graph.add((sdm_lab.Editor, RDFS.label, Literal("Editor")))

        #Author
        graph.add((sdm_lab.Reviewer, RDFS.subClassOf, sdm_lab.Author))
        graph.add((sdm_lab.Reviewer, RDFS.label, Literal("Reviewer")))

    #Function to save tbox in turtle file
    def write_tbox(self,graph,filename):
        graph.serialize(destination=f'../{filename}.ttl', format="ttl")
```

```python
#Create TBox for B1 
b1_RendinaWu = App()
graph = Graph()
b1_RendinaWu.tbox(graph)
b1_RendinaWu.write_tbox(graph,"TBox")
```
